Does Online Advertising Improve Sales?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind
```

Simulated Sales Data

```python
np.random.seed(42)
sales_before_ads = np.random.normal(loc=200, scale=20, size=30)
sales_after_ads = np.random.normal(loc=230, scale=25, size=30)
```

 Hypothesis Testing

```python
t_stat, p_value = ttest_ind(sales_after_ads, sales_before_ads)

alpha = 0.05
result = " Ads significantly improved sales" if p_value < alpha else " Ads had no significant effect"

print("t-statistic:", t_stat)
print("p-value:", p_value)
print("Result:", result)
```

Visualization

```python
df = pd.DataFrame({
    "Sales": np.concatenate([sales_before_ads, sales_after_ads]),
    "Group": ["Before Ads"]*len(sales_before_ads) + ["After Ads"]*len(sales_after_ads)
})
plt.figure(figsize=(8,5))
sns.boxplot(x="Group", y="Sales", data=df, hue="Group", palette="Set3", legend=False)
sns.swarmplot(x="Group", y="Sales", data=df, color=".25")
plt.title(f"Effect of Ads on Sales\n{result} (p={p_value:.4f})", fontsize=14)
plt.ylabel("Sales Amount (in thousands)")
plt.show()
```
