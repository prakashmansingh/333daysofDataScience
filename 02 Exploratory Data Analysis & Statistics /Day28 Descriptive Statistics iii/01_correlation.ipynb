```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

 Generate synthetic dataset

```python

np.random.seed(42)
data = {
    "Temperature": np.random.randint(15, 40, 50),  
    "Humidity": np.random.randint(30, 90, 50),     
    "Electricity_Consumption": np.random.randint(200, 600, 50) 
}
df = pd.DataFrame(data)
```

Compute Correlations

```python
#person correlation
pearson_corr = df.corr(method='pearson')
print("\nPearson Correlation:\n", pearson_corr)

#spearman correlation
spearman_corr = df.corr(method='spearman')
print("\nSpearman Correlation:\n", spearman_corr)
```

Visualize Pearson Correlations 

```python
plt.figure(figsize=(6,4))
sns.heatmap(pearson_corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Pearson Correlation Heatmap")
plt.show()


```

Visualize Spearman Correlations 

```python
plt.figure(figsize=(6,4))
sns.heatmap(spearman_corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Spearman Correlation Heatmap")
plt.show()
```

Scatter plot with Regression line

```python
plt.figure(figsize=(6,4))
sns.regplot(x="Temperature", y="Electricity_Consumption", data=df, scatter_kws={'alpha':0.7})
plt.title("Temperature vs Electricity Consumption")
plt.xlabel("Temperature (C)")
plt.ylabel("Electricity Consumption (kWh)")
plt.show()
```
