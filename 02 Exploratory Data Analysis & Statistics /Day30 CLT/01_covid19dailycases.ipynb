```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Generate the random data 

```python
np.random.seed(42)  
covid_cases = np.random.exponential(scale=200, size=1000) 
```

Plot original data distribution

```python

sns.histplot(covid_cases, bins=30, kde=True, color="red")
plt.title("Original Distribution of Daily COVID-19 Cases (Skewed)")
plt.xlabel("Daily Cases")
plt.ylabel("Frequency")
plt.show()
```

Apply Central Limit Theorem with Sampling

```python
sample_size = 50      
num_samples = 1000    
sample_means = []
for _ in range(num_samples):
    sample = np.random.choice(covid_cases, size=sample_size, replace=True)
    sample_means.append(sample.mean())
```

Plot Distribution of Sample Means

```python
plt.figure(figsize=(8,4))
sns.histplot(sample_means, bins=30, kde=True, color="blue")
plt.title(f"Distribution of Sample Means (n={sample_size}, {num_samples} samples)")
plt.xlabel("Sample Mean of Daily Cases")
plt.ylabel("Frequency")
plt.show()
```

Compare with Original

```python
plt.figure(figsize=(12,5))


# for Original distribution
plt.subplot(1,2,1)
sns.histplot(covid_cases, bins=30, kde=True, color="red")
plt.title("Original COVID-19 Cases (Skewed)")
plt.xlabel("Daily Cases")



# for Sample means distribution
plt.subplot(1,2,2)
sns.histplot(sample_means, bins=30, kde=True, color="blue")
plt.title("Sample Means (Approx Normal)")
plt.xlabel("Sample Mean of Cases")

plt.tight_layout()
plt.show()

```
