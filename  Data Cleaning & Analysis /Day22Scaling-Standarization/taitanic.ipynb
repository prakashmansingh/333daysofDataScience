```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

titanic = pd.read_csv("titanic.csv")
features = titanic[["Age", "Fare"]].dropna()
scaler = StandardScaler()
scaled = scaler.fit_transform(features)
titanic_standard = pd.DataFrame(scaled, columns=["Age_standardized", "Fare_standardized"])
print("Before Standardization:\n", features.describe())
```

```python
print("\nAfter Standardization:\n", titanic_standard.describe())
```

```python
#visualization
import matplotlib.pyplot as plt
fig , ax = plt.subplots(1, 2, figsize=(8, 4))
# Plotting the original features
ax[0].scatter(features["Age"], features["Fare"] )
ax[0].set_title("Original Features: Age vs Fare")
ax[0].set_xlabel("Age")
ax[0].set_ylabel("Fare")
# Plotting the standardized features
ax[1].scatter(titanic_standard["Age_standardized"], titanic_standard["Fare_standardized"])
ax[1].set_title("Standardized Features: Age vs Fare")
ax[1].set_xlabel("Age Standardized")
ax[1].set_ylabel("Fare Standardized")
plt.tight_layout()
plt.show()
```
